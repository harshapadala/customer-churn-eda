Import Libraries

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
sns.set_style("whitegrid")
```

Load Dataset

```python

df = pd.read_csv('/content/train.csv')

print("✅ Dataset Loaded")
print("Shape:", df.shape)

df.head()
```

Basic Understanding

```python

print("\nColumns:\n", df.columns.tolist())

print("\nData Types:")
display(df.dtypes)

print("\nSummary Statistics:")
display(df.describe())
```

Missing Values Analysis

```python

missing = df.isnull().sum().sort_values(ascending=False)
missing_percent = (missing / len(df)) * 100

missing_df = pd.DataFrame({
    'Missing Count': missing,
    'Missing %': missing_percent
})

missing_df[missing_df['Missing Count'] > 0].head(20)
```

Duplicate Check

```python

duplicates = df.duplicated().sum()
print(f"Duplicate rows: {duplicates}")
```

Target Variable Analysis

```python

plt.figure(figsize=(6,4))
sns.histplot(df['SalePrice'], kde=True)
plt.title("SalePrice Distribution")
plt.show()

print("Skewness:", df['SalePrice'].skew())
```

Log Transformation Check

```python

plt.figure(figsize=(6,4))
sns.histplot(np.log1p(df['SalePrice']), kde=True)
plt.title("Log Transformed SalePrice")
plt.show()
```

Numerical Features vs Price

```python

num_cols = df.select_dtypes(exclude='object').columns.tolist()
num_cols.remove('SalePrice')

top_num = num_cols[:6]  # first few important ones

for col in top_num:
    plt.figure(figsize=(5,4))
    sns.scatterplot(x=df[col], y=df['SalePrice'])
    plt.title(f"{col} vs SalePrice")
    plt.show()
```

Correlation Heatmap

```python

plt.figure(figsize=(12,8))
corr = df.corr(numeric_only=True)

# top correlated features with price
top_corr = corr['SalePrice'].sort_values(ascending=False).head(15).index

sns.heatmap(df[top_corr].corr(), annot=True, cmap='coolwarm')
plt.title("Top Features Correlated with SalePrice")
plt.show()
```

Categorical Feature Analysis

```python

cat_cols = df.select_dtypes(include='object').columns.tolist()

important_cats = cat_cols[:5]

for col in important_cats:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=df[col], y=df['SalePrice'])
    plt.xticks(rotation=45)
    plt.title(f"{col} vs SalePrice")
    plt.show()
```

Outlier Detection

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df['SalePrice'])
plt.title("SalePrice Outliers")
plt.show()
```
